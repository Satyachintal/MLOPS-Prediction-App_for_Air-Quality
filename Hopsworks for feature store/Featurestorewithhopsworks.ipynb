**Creating Feature Groups**

```python
import hopsworks

project = hopsworks.login()

fs = project.get_feature_store()
```

```python
# Get or create the 'airquality_data' feature group
data_prediction_fg= fs.get_or_create_feature_group(
    name="airquality_data",
    version=1,
    description="data for air quality prediction.",
    primary_key=['AQI'],

)
```

```python
# Insert data into feature group
data_prediction_fg.insert(df_filtered)
```
